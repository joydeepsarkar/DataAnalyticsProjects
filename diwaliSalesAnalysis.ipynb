```python
# Exploratory Data Analysis
```

```python
df.columns
```

```python
ax = sbs.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
barplot = sbs.barplot(x ='Gender',y='Amount',data=sales_gen)
for bars in barplot.containers:
    ax.bar_label(bars,fmt = '%d')
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
# import matplotlib.pyplot as plt
# plt.ticklabel_format(style='plain', axis='y',useOffset=False)
# fmt = '%d' removes the scientific notations
sales_agebarplot = sbs.barplot(x ='Age Group',y='Amount',data=sales_age)
for bars in sales_agebarplot.containers:
    sales_agebarplot.bar_label(bars,fmt = '%d')
```

```python
df.columns
```

```python
# total orders from top 10 states
orders_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
orders_state
```

```python
order_state_barplot = sbs.barplot(data = orders_state, x = 'State', y="Orders")
sbs.set(rc = {'figure.figsize':(25,5)})

# labelling the graph
for bars in order_state_barplot.containers:
    order_state_barplot.bar_label(bars)
```

```python
# Total amount from top 10 states

amount_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
amount_state
```

```python
amount_state_barplot = sbs.barplot(data = amount_state, x = 'State', y='Amount')
sbs.set(rc = {'figure.figsize':(25,5)})

# labelling the graph
for bars in amount_state_barplot.containers:
    amount_state_barplot.bar_label(bars, fmt = '%d')
```

```python
# df.columns
ax = sbs.countplot(data = df,x = 'MaritalStatus')
for bars in ax.containers:
    ax.bar_label(bars)
sbs.set(rc = {'figure.figsize':(6,5)})
```

```python
maritalStatusAmount = df.groupby(['MaritalStatus','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
maritalStatusAmount
```

```python
maritalStatusAmount_barplot = sbs.barplot(data = maritalStatusAmount, x = 'MaritalStatus', y="Amount",hue='Gender')
sbs.set(rc = {'figure.figsize':(7,5)})

# labelling the graph
for bars in maritalStatusAmount_barplot.containers:
    maritalStatusAmount_barplot.bar_label(bars, fmt = '%d')
```

```python
ax = sbs.countplot(data = df,x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
sbs.set(rc = {'figure.figsize':(25,5)})
```
